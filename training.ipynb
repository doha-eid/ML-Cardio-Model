```python
#Import dataset
import pandas as pd
import numpy as np

data = pd.read_csv ( 'cardio_train.csv' , sep = ";" )
```

```python
data
```

```python
# for i in range ( 0 , len(data['age']) ):
data['age'] = data['age'] / 360
```

```python
# data['age']
data
```

```python
# drop from the 13 column, cardio as it is the output , and the id as it has no change on the output
X = data.drop(['cardio','id','height'], axis=1)
y = data['cardio']
X 
# print the first two rows
# print(X.head(2))
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

# sns.set(rc = {'figure.figsize':(15,8)})
# sns.countplot(x='age',hue='cardio',data=data,palette='colorblind',edgecolor=sns.color_palette('dark',n_colors=1))
```

```python
#Splitting to training and testing
from sklearn.model_selection import train_test_split
# split the data into 70% train, 30% test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
#Train the model using Multi-Layer perception
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier( activation='relu')
mlp
```

```python
X_train.values
```

```python
#Training
# fit 70% of the data to do the train
mlp.fit(X_train,y_train)
```

```python
# get the last row in the data
X_test[-1:]
```

```python
#Testing
# predict the 30% of the data set as a test for the training phase on 70% of the data
pred = mlp.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
# a percentage to hoe accurate the model test the data
accuracy_score(y_test, pred)
# the model is good cause acc = 72.2 %
```
